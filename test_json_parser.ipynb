```python
def json_parser(filename, mode="valid"):
    data = json.load( open("val_and_test.json") )
    data = data[mode]
    n = len(data["experiment"])
    
    df = pd.DataFrame(columns=["exp", "run", "detector", "event"])
    
    for i in range(n):
        exp = data["experiment"][i]
        run = data["run"][i]
        det = data["detector"][i]
        events = data["events"][i].split(":")
        if len(events) == 3:
            events = list(range( int(events[0]), int(events[2]), int(events[1]) ))
        elif len(events) == 2:
            if len(events[0]) > 0 and len(events[1]) > 0:
                events = list(range( int(events[0]), int(events[1]) ))
            else:
                events = list(range(nEvents(str(exp), str(run), str(det))))
        elif len(events) == 1:
            events = [int(events[0])]
        my_data = [ [exp, run, det, e] for e in events ]
#         print(my_data)
        my_df = pd.DataFrame(my_data, columns=["exp", "run", "detector", "event"])
#         print(my_df)
        df = pd.concat( (df, my_df), axis=0 )
    
    return df
        
```

```python
df = json_parser("val_and_test.json", mode="valid")
df
```
